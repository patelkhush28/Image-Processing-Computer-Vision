```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def filter(img,filter):
    new_img=np.zeros(img.shape)
    for i in range (1,img.shape[0]-1):
        for j in range (1,img.shape[1]-1):
            new_img[i][j]=img[i-1][j-1]*filter[0][0]+img[i-1][j]*filter[0][1]+img[i-1][j+1]*filter[0][2]+img[i][j-1]*filter[1][0]+img[i][j]*filter[1][1]+img[i][j+1]*filter[1][2]+img[i+1][j-1]*filter[2][0]+img[i+1][j]*filter[2][1]+img[i+1][j+1]*filter[2][2]
            new_img=new_img.astype(np.uint8)
    return new_img
```

```python
image=cv2.imread('download (1).png',0)
plt.imshow(image,cmap='gray')
```

```python
p_x=[[-1,-1,-1],[0,0,0],[1,1,1]]
```

```python
p_y=np.array([[-1,0,1],
                  [-1,0,1],
                  [-1,0,1]])
```

```python
new_img1=filter(image,p_x)
plt.imshow(new_img1,cmap='gray')
```

```python
new_img2=filter(image,p_y)
plt.imshow(new_img2,cmap='gray')
```

```python
new_img3=filter(image,p_x+p_y)
plt.imshow(new_img3,cmap='gray')
```

```python
s_x=np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])

s_y=np.array([[-1,0,1],
                  [-2,0,2],
                  [-1,0,1]])
```

```python
new_imgs1=filter(image,s_x)
plt.imshow(new_imgs1,cmap='gray')
```

```python
new_imgs2=filter(image,s_y)
plt.imshow(new_imgs2,cmap='gray')
```

```python
new_imgs3=filter(image,s_x+s_y)
plt.imshow(new_imgs3,cmap='gray')
```
