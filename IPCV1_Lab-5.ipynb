```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def filter(img,filter):
    new_img=np.zeros(img.shape)
    for i in range (1,img.shape[0]-1):
        for j in range (1,img.shape[1]-1):
            new_img[i][j]=img[i-1][j-1]*filter[0][0]+img[i-1][j]*filter[0][1]+img[i-1][j+1]*filter[0][2]+img[i][j-1]*filter[1][0]+img[i][j]*filter[1][1]+img[i][j+1]*filter[1][2]+img[i+1][j-1]*filter[2][0]+img[i+1][j]*filter[2][1]+img[i+1][j+1]*filter[2][2]
            #new_img=new_img.astype(np.uint8)
    return new_img
```

```python
beach=cv2.imread('link.png',0)
plt.imshow(beach,cmap='gray')
```

```python
hpf=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
A=1
hbf=np.array([[-1,-1,-1],[-1,A+8,-1],[-1,-1,-1]])
```

```python
new_img1=filter(beach,hpf)
cv2.imshow('hpf',new_img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(new_img1,cmap='gray')
```

```python
new_img2=filter(beach,hbf)
plt.imshow(new_img2,cmap='gray')
```
