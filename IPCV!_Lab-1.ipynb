```python
import cv2
import matplotlib.pyplot as plt
```

```python
beach=cv2.imread('beach.jpg',0)
cv2.imshow("Beach",beach)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
land=cv2.imread('land.jpg',0)
cv2.imshow("land",land)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
beach=cv2.cvtColor(beach,cv2.COLOR_BGR2GRAY)
land=cv2.cvtColor(land,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Beach",beach)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("Gray land",land)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(beach)
plt.imshow(land)
```

### Arithmatic Operations

```python
land.resize(beach.shape)
```

```python
land.shape
beach.shape
```

```python
added_img=cv2.add(beach,land)
cv2.imshow("Added Image",added_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
sub_img=cv2.subtract(beach,land)
cv2.imshow("Subtracted Image",sub_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Logical Operators

```python
and_img=cv2.bitwise_and(beach,land)
cv2.imshow("and",and_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
or_img=cv2.bitwise_or(beach,land)
cv2.imshow("or",or_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
xor_img=cv2.bitwise_xor(beach,land)
cv2.imshow("xor",xor_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
not_img=cv2.bitwise_not(beach)
cv2.imshow("not",not_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
